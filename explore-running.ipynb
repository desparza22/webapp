```python
import docker

client = docker.from_env()
```

```python
image_tag = "ghcr.io/desparza22/packages-for-remote-exec/test-image:1.0"
```

```python
image = client.images.pull(image_tag)
print(f"pulled image {image.tags}")
```

```python
client.containers.run(
    image=image_tag,
    detach=False,       # True if you want to manage container manually
    remove=True,        # remove container when it exits
    stdout=True,
    stderr=True,
)
```

```python
container = client.containers.run(
    image=image_tag,
    detach=True,       # True if you want to manage container manually
    remove=False,        # remove container when it exits
    stdout=True,
    stderr=True,
)
```

```python
container.logs()
```

```python
container.remove()
```
